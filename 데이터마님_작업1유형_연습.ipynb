```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/youtube.csv",index_col=0)
df.head()

# Q1. 인기동영상 제작횟수가 많은 채널 상위 10개명을 출력하라 (날짜기준, 중복포함)
list(df.loc[df['channelId'].isin(df['channelId'].value_counts().head(10).index)]['channelTitle'].unique())

# Q2. 논란으로 인기동영상이 된 케이스를 확인하고 싶다. dislikes수가 like 수보다 높은 동영상을 제작한 채널을 모두 출력하z라
list(df.loc[df['dislikes'] > df['likes']]['channelTitle'].unique())

# Q3. 채널명을 바꾼 케이스가 있는지 확인하고 싶다. channelId의 경우 고유값이므로 이를 통해 채널명을 한번이라도 바꾼 채널의 갯수를 구하여라
change = df[['channelId', 'channelTitle']].drop_duplicates()['channelId'].value_counts()
len(change[change > 1])

# Q4. 일요일에 인기있었던 영상들중 가장많은 영상 종류(categoryId)는 무엇인가?
# 일요일 == 6
# 방법 1
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
df['new_trending_date2'] = df['trending_date2'].dt.weekday
df.loc[df['new_trending_date2'] == 6].sort_values('likes', ascending = True)['categoryId'].values[0]

# 방법 2
df['trending_date2'] = pd.to_datetime(df['trending_date2'])
answer =df.loc[df['trending_date2'].dt.day_name() =='Sunday']['categoryId'].value_counts().index[0]
# print(answer)

# Q5. 각 요일별 인기 영상들의 categoryId는 각각 몇개 씩인지 하나의 데이터 프레임으로 표현하라
dff = df.groupby([df['trending_date2'].dt.day_name(), 'categoryId'], as_index=False).size()
dff.pivot_table(index = 'categoryId', columns = dff['trending_date2'])

# Q6. 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다. viewcount대비 댓글수가 가장 높은 영상을 확인하라 (view_count값이 0인 경우는 제외한다)
target2= df.loc[df['view_count']!=0] # view_count가 0이 아닌 것만 추출 -> 0인 것은 제거
t = target2.copy()
t['ratio'] = (target2['comment_count']/target2['view_count']).dropna()
t.sort_values('ratio', ascending=False).iloc[0]['title']

# Q7. 댓글의 수로 (comment_count) 영상 반응에 대한 판단을 할 수 있다.viewcount대비 댓글수가 가장 낮은 영상을 확인하라 (view_counts, ratio값이 0인경우는 제외한다.)
tar = df.loc[df['view_count'] != 0]
t = tar.copy()
t['ratio'] = (target2['comment_count']/target2['view_count']).dropna()
t['ratio'] = t['ratio'][t['ratio'] != 0]
t.sort_values('ratio', ascending = True).iloc[0]['title']

# Q8. like 대비 dislike의 수가 가장 적은 영상은 무엇인가? (like, dislike 값이 0인경우는 제외한다)
tar = df.loc[(df['likes'] != 0) & (df['dislikes'] != 0)]
t = tar.copy()
t['ratio'] = (tar['dislikes'] / tar['likes']).dropna()
t.sort_values('ratio').iloc[0]['title']

# Q9. 가장많은 트렌드 영상을 제작한 채널의 이름은 무엇인가? (날짜기준, 중복포함)
# 방법 1
df.loc[df['channelId'] == df['channelId'].value_counts().index[0]].sort_values('trending_date2', ascending = False).iloc[0]['channelTitle']
# 방법 2
df.loc[df['channelId'] == df['channelId'].value_counts().index[0]]['channelTitle'].unique()[0]

# Q10. 20회(20일)이상 인기동영상 리스트에 포함된 동영상의 숫자는?
(df[['title', 'channelId']].value_counts() >= 20).sum(axis = 0)
```

```python
# Q11. 각 데이터의 ‘ct’컬럼을 시간으로 인식할수 있게 datatype을 변경하고 video 데이터의 videoname의 각 value 마다 몇개의 데이터씩 가지고 있는지 확인하라
channel['ct'] = pd.to_datetime(channel['ct'])
video['ct'] = pd.to_datetime(video['ct'])
# channel.info()
# video.info()

video['videoname'].value_counts()

# Q12. 수집된 각 video의 가장 최신화 된 날짜의 viewcount값을 출력하라
video.sort_values(['videoname','ct']).drop_duplicates('videoname',keep='last')[['viewcnt','videoname','ct']].reset_index(drop = True)
#video.sort_values(['videoname','ct']).drop_duplicates('videoname',keep='last')[['viewcnt','videoname','ct']].reset_index(drop=True)

# Q13. Channel 데이터중 2021-10-03일 이후 각 채널의 처음 기록 됐던 구독자 수(subcnt)를 출력하라
channel[channel['ct'] >= pd.to_datetime('2021-10-03')].sort_values(['ct','channelname']).drop_duplicates('channelname',keep='first')[['channelname', 'subcnt']].reset_index(drop = True)
# target = channel[channel['ct'] >= pd.to_datetime('2021-10-03')].sort_values(['ct','channelname']).drop_duplicates('channelname')
# answer = target[['channelname','subcnt']].reset_index(drop=True)

# Q14. 각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라
# channel[(channel['ct'] >= pd.to_datetime('2021-10-03 03:00:00')) & (channel['ct'] <= pd.to_datetime('2021-11-01 15:00:00'))].reset_index(drop=True)
end = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') =='2021-11-01 15']
start = channel.loc[channel.ct.dt.strftime('%Y-%m-%d %H') =='2021-10-03 03']

end_df = end[['channelname','subcnt']].reset_index(drop=True)
start_df = start[['channelname','subcnt']].reset_index(drop=True)

end_df.columns = ['channelname','end_sub']
start_df.columns = ['channelname','start_sub']

tt = pd.merge(start_df,end_df)
tt['del'] = tt['end_sub'] - tt['start_sub']
result = tt[['channelname','del']]
result

# Q15. 각 비디오는 10분 간격으로 구독자수, 좋아요, 싫어요수, 댓글수가 수집된것으로 알려졌다.
# 공범 EP1의 비디오정보 데이터중 수집간격이 5분 이하, 20분이상인 데이터 구간( 해당 시점 전,후) 의 시각을 모두 출력하라
# timedelta : 두 날짜의 차이를 알기 위한 함수
# diff : 한 객체 내에서 열과 열 / 행과 행의 차이를 출력하는 메서드
import datetime

dff = video.loc[video['videoname'].str.contains('공범 EP1')].sort_values('ct').reset_index(drop = True)
dff[(dff['ct'].diff(1) <= datetime.timedelta(minutes = 5)) | (dff['ct'].diff(1) <= datetime.timedelta(minutes = 20))]

# Q16. 각 에피소드의 시작날짜(년-월-일)를 에피소드 이름과 묶어 데이터 프레임으로 만들고 출력하라
start_date = video.sort_values(['ct','videoname']).drop_duplicates('videoname')[['ct','videoname']]
start_date['date'] = start_date['ct'].dt.date
answer = start_date[['date','videoname']]
# display(answer)

# Q17. “공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.
# 공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라
dff = video[video['videoname'].str.contains('공범')]
dff[dff['ct'].dt.hour == 21].drop_duplicates('videoname').sort_values('viewcnt', ascending = False)[['videoname', 'viewcnt', 'ct']].reset_index(drop = True)

# Q18. video 정보의 가장 최근 데이터들에서 각 에피소드의 싫어요/좋아요 비율을 ratio 컬럼으로 만들고 videoname, ratio로 구성된 데이터 프레임을 ratio를 오름차순으로 정렬하라
dff = video.sort_values('ct', ascending = False).drop_duplicates('videoname', keep='first')
dff['ratio'] = dff['dislikecnt'] / dff['likecnt']
result = dff[['videoname', 'ratio']].sort_values('ratio').reset_index(drop = True)
result

# Q19. 2021-11-01 00:00:00 ~ 15:00:00까지 각 에피소드별 viewcnt의 증가량을 데이터 프레임으로 만드시오
# 제시해준 시간을 datetime 형태로 변환
start = pd.to_datetime("2021-11-01 00:00:00")
end = pd.to_datetime("2021-11-01 15:00:00")

# 조건에 맞는 컬럼 추출
target = video.loc[(video["ct"] >= start) & (video['ct'] <= end)].reset_index(drop=True)

# 증가량 차이 구하는 함수
def check(x):
    result = max(x) - min(x)
    return result

answer = target[['videoname','viewcnt']].groupby("videoname").agg(check)
answer

# Q20. video 데이터 중에서 중복되는 데이터가 존재한다. 중복되는 각 데이터의 시간대와 videoname 을 구하여라
answer  = video[video.index.isin(set(video.index) -  set(video.drop_duplicates().index))]
result = answer[['videoname','ct']]
display(result)
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/bicycle/seoul_bi.csv')
df.head()
```

```python
import datetime

# Q31. 대여일자별 데이터의 수를 데이터프레임으로 출력하고,
dff = df['대여일자'].value_counts().T.to_frame().sort_index()
dff

# 가장 많은 데이터가 있는 날짜를 출력하라
# print(dff.sort_values('대여일자', ascending = False).index[0])

# Q32. 각 일자의 요일을 표기하고 (‘Monday’ ~’Sunday’) ‘day_name’컬럼을 추가하고 이를 이용하여 각 요일별 이용 횟수의 총합을 데이터 프레임으로 출력하라
df['대여일자'] = pd.to_datetime(df['대여일자'])
df['day_name'] = df['대여일자'].dt.day_name()
df['day_name'].value_counts().to_frame()

# Q33. 각 요일별 가장 많이 이용한 대여소의 이용횟수와 대여소 번호를 데이터 프레임으로 출력하라
dff = df.groupby(['day_name', '대여소번호']).size().to_frame('size').reset_index(drop = False).sort_values(['day_name', 'size'], ascending = False).drop_duplicates('day_name', keep = 'first')
# to_frame('size') : 새로운 컬럼명을 생성하고 싶을 때
dff.reset_index(drop = True)

# Q34. 나이대별 대여구분 코드의 (일일권/전체횟수) 비율을 구한 후 가장 높은 비율을 가지는 나이대를 확인하라.
# 일일권의 경우 일일권 과 일일권(비회원)을 모두 포함하라
chungki = df[(df['대여구분코드'] == '일일권') | (df['대여구분코드'] == '일일권(비회원)')].groupby('연령대코드').size()
result = chungki / df.groupby('연령대코드').size()
result = result.sort_values(ascending = False)
# print(result)
# print('max ratio age', result.index[0])

# Q35. 연령대별 평균 이동거리를 구하여라
df.groupby(['연령대코드'])['이동거리'].mean()

# Q36. 연령대 코드가 20대인 데이터를 추출하고,
# 이동거리값이 추출한 데이터의 이동거리값의 평균 이상인 데이터를 추출한다.
# 최종 추출된 데이터를 대여일자, 대여소 번호 순서로 내림차순 정렬 후
# 1행부터 200행까지의 탄소량의 평균을 소숫점 3째 자리까지 구하여라
df20 = df[df['연령대코드'] == '20대']
dfleng = df20[df20['이동거리'] >= df20['이동거리'].mean()]
dff = dfleng.sort_values(['대여일자', '대여소번호'], ascending = False).iloc[0:200]
dff['탄소량'] = dff['탄소량'].astype('float')
round(dff['탄소량'].mean(), 3)

# Q37. 6월 7일 ~10대의 “이용건수”의 중앙값은?
df_date = df[(df['대여일자'].dt.month == 6) & (df['대여일자'].dt.day == 7)]
df_date[df_date['연령대코드'] == '~10대']['이용건수'].median()

# Q38. 평일 (월~금) 출근 시간대(오전 6,7,8시)의 대여소별 이용 횟수를 구해서 데이터 프레임 형태로 표현한 후
# 각 대여시간별 이용 횟수의 상위 3개 대여소와 이용횟수를 출력하라
df_week = df[(df['day_name'] != 'Saturday') & (df['day_name'] != 'Sunday')]
df_time = df_week[(df_week['대여시간'] == 6) | (df_week['대여시간'] == 7) | (df_week['대여시간'] == 8)]
df_sort = df_time.groupby(['대여시간', '대여소번호']).size().to_frame('이용 횟수')
df_sort.sort_values(['대여시간', '이용 횟수'], ascending = False).apply(lambda x: x.nlargest(3, '이용 횟수'))
```

```python
pd.set_option('display.max_columns',50)
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/nba/nba.csv",encoding='latin',sep=';')
df.head()

# Q104. Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라
round(df[df['Tm'] == 'TOR']['Age'].mean(), 4)

# Q105. Pos 컬럼은 포지션을 의미한다.
# 전체 선수 중 최소나이대의 선수들을 필터하고 그들 중 가장 많은 포지션은 무엇인지 확인하라
# print(df[df['Age'] == df['Age'].min()]['Pos'].value_counts().index[0])

# Q106. 선수들의 이름은 first_name+ 공백 + last_name으로 이루어져 있다. 가장 많은 first_name은 무엇이며 몇 회 발생하는지 확인하라
df['first_name'] = df['Player'].str.split(' ').str[0]
# print(df['first_name'].value_counts().index[0], df['first_name'].value_counts()[0])

# Q107. PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라
df.groupby('Pos')['PTS'].mean().sort_values(ascending = True)

# Q109. G컬럼은 참여한 경기의 숫자이다.
# 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라
df.sort_values(['Tm','G']).groupby('Tm').tail(1)['G'].mean()

# Q110. Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?
df[(df['Tm'] == 'MIA') & ((df['Pos'] == 'C') | (df['Pos'] == 'PF')) ]['MP'].mean()

# Q111. 전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?
df[df['G'] >= df['G'].mean() * 1.5]['3P'].mean()

# Q112. Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?
df_up = df[df['Age'] >= df['Age'].mean()]['G'].mean()
df_down = df[df['Age'] < df['Age'].mean()]['G'].mean()
abs(df_up - df_down)

# Q113. 평균나이가 가장 젊은 팀은 어디인가
df.groupby('Tm')['Age'].mean().sort_values().index[0]

# Q114. Pos그룹별 평균 MP값을 구하여라
df.groupby('Pos')['MP'].mean()
```
